```
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

import numpy as np
```

```
# Read data from file
df = pd.read_csv('epa-sea-level.csv')
df.info
```

```
df.head()
```

```
# Create scatter plot
fig, ax = plt.subplots(figsize = (15, 10))

ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```
# Create first line of best fit
slope, intercept, r_value, p_value, std_err  = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
print(slope, intercept)
```

```
years = np.asarray(list(range(1880, 2051, 1)))
ax.plot(years, intercept + slope * years)
fig
```

```
# Create second line of best fit
df2 = df.copy()
df2 = df2[df2['Year'] >= 2000]
slope2, intercept2, r_value2, p_value2, std_err2  = linregress(df2['Year'], df2['CSIRO Adjusted Sea Level'])
print(slope2, intercept2)
```

```
years2 = np.asarray(list(range(2000, 2051, 1)))
ax.plot(years2, intercept2 + slope2 * years2)
fig
```

```
# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```

```
def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')
    
    # Create scatter plot
    fig, ax = plt.subplots(figsize = (15, 10))

    ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])


    # Create first line of best fit
    slope, intercept, r_value, p_value, std_err  = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

    years = np.asarray(list(range(1880, 2051, 1)))
    ax.plot(years, intercept + slope * years)

    # Create second line of best fit
    df2 = df.copy()
    df2 = df2[df2['Year'] >= 2000]
    slope2, intercept2, r_value2, p_value2, std_err2  = linregress(df2['Year'], df2['CSIRO Adjusted Sea Level'])

    years2 = np.asarray(list(range(2000, 2051, 1)))
    ax.plot(years2, intercept2 + slope2 * years2)

    # Add labels and title
    ax.set_xlabel('Year')
    ax.set_ylabel('Sea Level (inches)')
    ax.set_title('Rise in Sea Level')

    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```
